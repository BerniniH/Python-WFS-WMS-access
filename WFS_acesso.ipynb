```python
from ipyleaflet import Map, WMSLayer, LayersControl, FullScreenControl
from owslib.wfs import WebFeatureService
import geopandas as gpd
```

```python

# Definir a URL do serviço WFS
wfs_pfg = "https://panorama.sipam.gov.br/geoserver/painel_do_fogo/wfs"

# Conectar ao serviço WFS
pfg = WebFeatureService(url=wfs_pfg, version='1.1.0')
```

```python

# Definir a URL do serviço WFS
wfs_bdq = "https://terrabrasilis.dpi.inpe.br/queimadas/geoserver/dados_abertos/wfs"

# Conectar ao serviço WFS
bdq = WebFeatureService(url=wfs_bdq, version='1.1.0')

list(bdq.contents)
```

```python
# Nome do layer que queremos acessar
layer_name1 = "painel_do_fogo:mv_evento_filtro_exportKML"
layer_name2 = "dados_abertos:areas_atencao_goes"
```

```python

# Criar um mapa interativo com ipyleaflet
m = Map(center=[-9, -63], zoom=8)

# Adicionar uma camada WMS ao mapa (caso disponível)
wms_pfg = WMSLayer(
    url=wfs_pfg.replace('wfs', 'wms'),
    layers=layer_name1,
    transparent=True,
    format='image/png',
)

wms_bdq = WMSLayer(
    url=wfs_bdq.replace('wfs', 'wms'),
    layers=layer_name2,
    transparent=True,
    format='image/png',
    attribution='Projeto Biomas BR - DescrEVE',
)

# Adicionar a camada ao mapa
m.add_layer(wms_pfg)
m.add_layer(wms_bdq)

# Adicionar controles ao mapa
m.add_control(LayersControl())
m.add_control(FullScreenControl())

# Exibir o mapa
m
```
